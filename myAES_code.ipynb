```python
#galois multiplication
from copy import copy
import numpy as np
def g_mtp(a,b):
    if b == 1:
        return a
    elif b == 2:
        if a & 0x80 == 0x80:
            return ((a<<1)^(0x1b))& 255
        else:
            return (a<<1)& 255
    else:
        if a & 0x80 == 0x80:
            return ((a<<1)^(0x1b)^a)&255
        else:
            return ((a<<1)^a)&255 

    
def col_trans(state):
    new_state = np.zeros([4,4],dtype=int)
    
    A = np.array([
            [2,3,1,1],
            [1,2,3,1],
            [1,1,2,3],
            [3,1,1,2]])
        
    for j in range(len(state)):
        
        col = copy(state[:,j])
        x0,x1,x2,x3 = 0,0,0,0
        
        for i in range(4):

            x0 ^= g_mtp(col[i],A[0][i]) 
            x1 ^= g_mtp(col[i],A[1][i]) 
            x2 ^= g_mtp(col[i],A[2][i]) 
            x3 ^= g_mtp(col[i],A[3][i])
            x = [x0,x1,x2,x3]
        new_state[:,j] = [int(y) for y in x]
#         print(hex(x0),hex(x1),hex(x2),hex(x3))  
    return new_state


    
```
